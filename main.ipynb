# t-sneの実装

**目的**  
次元削減手法であるPCAとt-sneを実装してみる。

## PCA

### 所感
参考のサイトと点対象？  
PCAは固有値分解した固有ベクトルをデータ点にかけて圧縮したものだから再現性はあるはずでは？

```python
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.decomposition import PCA

digits = datasets.load_digits()
```

```python
display(digits.data.shape)
display(digits.target.shape)
```

```python
X_reduced = PCA(n_components = 2).fit_transform(digits.data)

X_reduced.shape
```

```python
plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=digits.target, cmap="jet")
plt.colorbar()
```

## t-SNE

[scikit-learnの公式リファレンス](https://scikit-learn.org/stable/modules/generated/sklearn.manifold.TSNE.html)

### 所感
PCAと比べると確かにクラスタリングできている。しかし参考サイトとは結果が異なる。  
バージョン違いによるものか？


```python
from sklearn.manifold import TSNE

X_reduced = TSNE(n_components=2, random_state = 0).fit_transform(digits.data)

X_reduced.shape
```

```python
plt.scatter(X_reduced[:,0], X_reduced[:,1], c=digits.target, cmap="jet")
plt.colorbar()
```
